```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
df = pd.read_csv('../data/processed/parsed_fasta/disorder_pdb_proteins.csv')
df
```

```python
total_residues = df['seq_len'].sum()
print(f'Total residues: {total_residues}')
total_disordered = df['disorder_count'].sum()
print(f'Total disordered residues: {total_disordered}')
total_ordered = df['order_count'].sum()
print(f'Total ordered residues: {total_ordered}')

total_annotated_residues = total_disordered + total_ordered
print(f'disorder + order: {total_disordered + total_ordered}')

disorder_fraction = total_disordered / total_annotated_residues
order_fraction = total_ordered / total_annotated_residues

print(f'Disorder fraction: {disorder_fraction:.3f}')
print(f'Order fraction: {order_fraction:.3f}')

total_na = df['na_count'].sum()
print(f'Total NA residues: {total_na}')
```
